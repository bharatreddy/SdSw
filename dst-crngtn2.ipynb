```python
from numpy import *
import datetime
import os.path
from calendar import monthrange
from matplotlib.dates import MonthLocator, WeekdayLocator, DateFormatter
import matplotlib as mp
import numpy as np


# Time for some datetime operations
# Get the date and time values of start and end dates
start_date_plot = datetime.datetime(2012,1,1)
end_date_plot = datetime.date.today()

start_month=start_date_plot.month
end_months=(end_date_plot.year-start_date_plot.year)*12 + end_date_plot.month+1


dates_for_files=[datetime.datetime(year=yr, month=mn, day=1) for (yr, mn) in (
          ((m - 1) / 12 + start_date_plot.year, (m - 1) % 12 + 1) for m in range(start_month, end_months)
      )]
Mon_array =  []
Year_array = []

for myind in range( len( dates_for_files ) ) :
    Mon_array.append( dates_for_files[myind].month )
    Year_array.append( dates_for_files[myind].year )


# setup the file string to be  read
# we need a zero before 1-9 months for the filename
# We'll setup a string array for month
Mon_array_str = []

for m in range(len(Mon_array)):
    if Mon_array[m] <= 9 :
        Mon_array_str.append( str(0) + str(Mon_array[m]) )

    else :
        Mon_array_str.append( str( Mon_array[m] ) )



#This array is for converting Dst back into float
#We want floats because python NaNs are floats
dst_val = []

#We need the date/time array at the same time too..
date_dst_arr = []

# The first date value is the first hour of first day as we know..
date_dst_arr.append( datetime.datetime(start_date_plot.year, Mon_array[0], 1, 0) )


# Dst indices are plotted every hour...so our time delta is 1 hour
dst_time_del = datetime.timedelta(hours = 1)



# Loop through the files of each month
for mon_index in range(len(Mon_array_str)):

    File_str = "/sd-data/dst/"+str(Year_array[mon_index])+"/"+str(Year_array[mon_index])+Mon_array_str[mon_index]+"_dst.dat"
    

    # Check if the file exists
    check_file = os.path.isfile( File_str )

    # Proceed only if the file exists
    if ( check_file == True ) :

        # Read the data into an array
        f = file(File_str, "r")
        s = f.read()
        f.close()

        # We have data in int.
        # So we have our array in int too...
        dst_val_temp = fromstring( s, int16 )


        for index in range( len( dst_val_temp ) ):

         # Check if we are in the proper date of month
            checkdd = date_dst_arr[-1] + dst_time_del
            if (checkdd.month != Mon_array[ mon_index ] ) :
                break

            
        # We are converting each element into float
            dst_val.append( float( dst_val_temp[ index ] ) )

        # Get the corresponding time information
            date_dst_arr.append ( date_dst_arr[-1] + dst_time_del )

        # Check for absurd values and convert them to NaNs
            if ( abs( dst_val[ -1 ] ) >= 9000 ) :
                dst_val[ -1 ] =  float( 'nan' )

    else :
        print 'No file found : ', File_str



# We have an additional element in date_dst_arr array
# since we appended the date_dst_arr before the loop began
date_dst_arr = date_dst_arr[ 0:len( date_dst_arr )-1 ]


# Loop through each day in the array and
# Get the min Dst value in a day and store it in an array
day_dst_arr = []
mon_dst_arr = []
dst_day_min_val = []
hour_val_dst_min = []


for mon_index in range(len(Mon_array_str)):
    curr_month = Mon_array[mon_index]
    curr_year = Year_array[mon_index]

    Ndays_month = monthrange(curr_year, curr_month)

    for mons in range( 1, Ndays_month[1]+1 ):

        # Get all the 24 values in that day
        curr_day_hrs = [ i for i, x in enumerate( date_dst_arr ) if ( x.day == mons and x.month == curr_month and x.year == curr_year ) ]
        #Just store the first value of the search since the entire array has same values

        day_dst_arr.append(  date_dst_arr[ curr_day_hrs[0] ]  )

        mon_dst_arr.append(  date_dst_arr[ curr_day_hrs[0] ].month  )
        #Get the min dst value for the day
        dst_day_min_val.append( min( dst_val[x] for x in curr_day_hrs ) )

        # Get the hour when we find min dst
        for kk in range( len( curr_day_hrs ) ) :

            if dst_val[ curr_day_hrs[ kk ] ] == dst_day_min_val[-1] :
                hour_val_dst_min.append( date_dst_arr[ curr_day_hrs[ kk ] ].hour )
                break



dst_array_min = np.array( dst_day_min_val )
days_array_dst = np.array( day_dst_arr )
hour_val_dst_min = np.array( hour_val_dst_min )



# Num. Days in crngtn rotation
Crngtn_rtn = 27.2753


# Get the first carrington day in the cycle
Crngtn_2012_first = datetime.datetime( 2012, 1, 9 )
# A little wierd to add 0.9... days since that was how I got it online
# So making sure it is as accurate as possible
Crngtn_2012_first = Crngtn_2012_first + datetime.timedelta( days = 0.9416 )



NextCrngtn = Crngtn_2012_first + datetime.timedelta( days = Crngtn_rtn )

# Start getting the carrington dates/cycles and other stuff into arrays
Crngtn_cycle_num = []
Crngtn_dates = []
Crngtn_dates.append( Crngtn_2012_first )
Crngtn_cycle_num.append( 2119 )

print Crngtn_rtn, Crngtn_2012_first

for k in range( 1, 25 ) :

    Crngtn_dates.append( Crngtn_dates[-1] + datetime.timedelta( days = Crngtn_rtn ) )
    Crngtn_cycle_num.append( Crngtn_cycle_num[-1] + 1 )

    if ( Crngtn_dates[-1].year == 2013 and Crngtn_dates[-1].year == 2 ) :
        break


Crngtn_dates = Crngtn_dates[ 0:len( Crngtn_dates )-1 ]

dst_50nt_inds = [ j for j, y in enumerate( dst_array_min ) if (  y <= -50 ) ]
dst_50nT_arr = np.array(dst_array_min[ dst_50nt_inds ] )
date_dst_50nT_arr = np.array( days_array_dst[ dst_50nt_inds ] )



hour_dst_50nT_arr = np.array( hour_val_dst_min[ dst_50nt_inds ] )

date_pred_50nt_arr = []
date_2cycpred_50nt_arr = []
fake_cnst = []

for d in range( len( date_dst_50nT_arr ) ):
    date_pred_50nt_arr.append( date_dst_50nT_arr[d] + datetime.timedelta( days = Crngtn_rtn ) )
    date_2cycpred_50nt_arr.append( date_dst_50nT_arr[d] + datetime.timedelta( days = Crngtn_rtn*2 ) )
    fake_cnst.append( dst_50nT_arr[d] )

    

date_pred_50nt_arr = np.array( date_pred_50nt_arr )
date_2cycpred_50nt_arr = np.array( date_2cycpred_50nt_arr )
fake_cnst = np.array( fake_cnst )
#ax.xaxis.set_minor_locator(xtickMins_minor)


# Get to the plotting part...
fig = figure(figsize = ( 11, 8.5 ) )
ax = fig.gca()

date_st = datetime.datetime(2012,9,11)
date_end = datetime.datetime(2013,7,1)
xtickMins_minor = mp.dates.DayLocator(interval = 2)

ax.plot_date( day_dst_arr,dst_array_min,'b-', linewidth = 2 )
pr3, = ax.plot_date( date_2cycpred_50nt_arr, fake_cnst, 's', markerfacecolor='gray', label="Projected event (2 cycles) ", markersize = 5 )
pr1, = ax.plot_date( date_pred_50nt_arr, fake_cnst, '.', markerfacecolor='gray', label="Projected event (1 cycle) " , markersize = 10)
ac1, = ax.plot_date( date_dst_50nT_arr, dst_50nT_arr, 'rx' , label = "Dst Event (< -50 nT) ", markersize = 12 )

#Join the actual and predicted points

for dt50, dtp1, dtp2, ds50, fc in zip( date_dst_50nT_arr, date_pred_50nt_arr, date_2cycpred_50nt_arr, dst_50nT_arr, fake_cnst ) :
    ax.plot_date( [ dt50, dtp1 ], [ ds50, fc ], ':', color = 'gray', linewidth = 1. )
    ax.plot_date( [ dtp1, dtp2 ], [ fc, fc ], ':', color = 'gray', linewidth = 1. )

    
# If the predicted days have a storm nearby actually happening mark them with a small red cross
for dpp1 in range( len( date_pred_50nt_arr ) ) :
    for ds11 in range( len ( date_dst_50nT_arr ) ) :
        if ( date_pred_50nt_arr[ dpp1 ].year == date_dst_50nT_arr[ ds11 ].year and date_pred_50nt_arr[ dpp1 ].month == date_dst_50nT_arr[ ds11 ].month and math.fabs( date_pred_50nt_arr[ dpp1 ].day - date_dst_50nT_arr[ ds11 ].day ) <= 1 ):
            p1c_rlzd, = ax.plot_date( date_pred_50nt_arr[ dpp1 ], fake_cnst[ dpp1 ], 'rx', markersize = 6.5 )
    

    

for dpp2 in range( len( date_2cycpred_50nt_arr ) ) :
    for ds12 in range( len ( date_dst_50nT_arr ) ) :
        if ( date_2cycpred_50nt_arr[ dpp2 ].year == date_dst_50nT_arr[ ds12 ].year and date_2cycpred_50nt_arr[ dpp2 ].month == date_dst_50nT_arr[ ds12 ].month and math.fabs( date_2cycpred_50nt_arr[ dpp2 ].day - date_dst_50nT_arr[ ds12 ].day ) <= 1 ):
            p2c_rlzd, = ax.plot_date( date_2cycpred_50nt_arr[ dpp2 ], fake_cnst[ dpp2 ], 'rx', markersize = 6.5 )


leg_labels = [ "Dst Event (< -50 nT) ", "Projected event (1 cycle)", "Projected event (2 cycles)", "Realized event (1 cycle)", "Realized event (2 cycles)" ]
#handles, labels = ax.get_legend_handles_labels()
fontProp = mp.font_manager.FontProperties( size = 7.5 )
#ax.legend(handles[::-1], labels[::-1], bbox_to_anchor=(1., 0.95), loc = 1, prop = fontProp)
ax.legend([ ac1, pr1, pr3, ( pr1, p1c_rlzd ), ( pr3, p2c_rlzd ) ], leg_labels, bbox_to_anchor=(1., 0.97), loc = 1, prop = fontProp)


# Find the  last date with dst data we plotted and annotate it
for d2 in range( len( days_array_dst ) ):
    if not math.isnan( dst_array_min[ d2 ] ) :
        last_good_dst_date = days_array_dst[ d2 ]
        good_dst_last = dst_array_min[ d2 ]
       
# Annotate the date to show the date of latest data   
ax.annotate( str( last_good_dst_date.month )+'/'+ str( last_good_dst_date.day ), xy =( last_good_dst_date, good_dst_last ), size = 8 )

# Annotate dates predicted for the coming cycle..
plabels_1cyc = [ str( date_pred_50nt_arr[-1].month )+'/'+str( date_pred_50nt_arr[-1].day ) ]

plabels_2cyc = [str( date_2cycpred_50nt_arr[-3].month )+'/'+str( date_2cycpred_50nt_arr[-3].day ), str( date_2cycpred_50nt_arr[-2].month )+'/'+str( date_2cycpred_50nt_arr[-2].day ), str( date_2cycpred_50nt_arr[-1].month )+'/'+str( date_2cycpred_50nt_arr[-1].day ) ]


plabels_xdate_1cyc = [ date_pred_50nt_arr[-1].date() ]
plabels_xdate_2cyc = [ date_2cycpred_50nt_arr[-3].date(), date_2cycpred_50nt_arr[-2].date(), date_2cycpred_50nt_arr[-1].date() ]
plabels_ydst_1cyc = [ fake_cnst[-1] ]
plabels_ydst_2cyc = [ fake_cnst[-3], fake_cnst[-2], fake_cnst[-1] ]
ha_pts = ['left', 'right', 'left']
va_pts = ['top', 'bottom', 'bottom' ]



for l, x, y, h, v in zip( plabels_2cyc, plabels_xdate_2cyc, plabels_ydst_2cyc, ha_pts, va_pts ) :
    ax.annotate(
    l,
    xy =( x, y ), xytext = (-25,25),
    textcoords = 'offset points', ha = h, va = v, size = 10,
    bbox = dict( boxstyle = 'round, pad = 0.1', fc = 'yellow', alpha = 0.25 ),
    arrowprops = dict( arrowstyle ='-', connectionstyle = 'arc3, rad = 0' ) )

for l, x, y, h, v in zip( plabels_1cyc, plabels_xdate_1cyc, plabels_ydst_1cyc, ha_pts, va_pts ) :
    ax.annotate(
    l,
    xy =( x, y ), xytext = (-10,10),
    textcoords = 'offset points', ha = h, va = v, size = 10,
    bbox = dict( boxstyle = 'round, pad = 0.1', fc = 'yellow', alpha = 0.25 ),
    arrowprops = dict( arrowstyle ='-', connectionstyle = 'arc3, rad = 0' ) )
   
   
# we need to annotate the date with highest Dst deviation in each carrington cycle as red
# select the dates which satisfy this criteria and make appropriate changes in the next part
# where we annotate dates...

dst_50_mark_check = []
date_dst_50_mark_check = []
date_dst_50_mark_red = []

for ccr_dates in range( len( Crngtn_dates ) ) :
    for dd50, dstt50 in zip( date_dst_50nT_arr, dst_50nT_arr) :
        if ( dd50 >=Crngtn_dates[ ccr_dates ] and dd50 <= Crngtn_dates[ ccr_dates + 1 ] ):
            dst_50_mark_check.append( dstt50 )
            date_dst_50_mark_check.append( dd50 )
    if len(date_dst_50_mark_check) > 0 :
        date_red_mark_inds = [ k for k, yy in enumerate( dst_50_mark_check ) if (  yy == min( dst_50_mark_check ) ) ]
        date_dst_50_mark_red.append( date_dst_50_mark_check[ date_red_mark_inds[ 0 ] ] )
       
   
    date_dst_50_mark_check = []
    dst_50_mark_check = []


# Annotate dates in the cycle where we max deviations in dst
count_add_pos = 0
old_cr_xpos = 0
new_cr_pos = 0
for a, b, c in zip( date_dst_50nT_arr[ ::-1 ], dst_50nT_arr[ ::-1 ],hour_dst_50nT_arr[ ::-1 ] ) :
    str_curr =str(a.month)+'/'+str(a.day)+"-"+str(c)+"UT"


    for count_cr_dates in range( len( Crngtn_dates ) ) :
        if ( a >= Crngtn_dates[ count_cr_dates ] and a <= Crngtn_dates[ count_cr_dates + 1 ] ) :
            xpos_dates = Crngtn_dates[ count_cr_dates ] + datetime.timedelta( days = 3 )
            new_cr_pos = count_cr_dates
            break


    if ( new_cr_pos == old_cr_xpos ) :
        count_add_pos = count_add_pos + 4
    else :
        old_cr_xpos = new_cr_pos
        count_add_pos = 0
    
    check_for_red_dates = 0   
    for lll in date_dst_50_mark_red :
        if ( lll == a ) :
            check_for_red_dates = 1
            break
           
    if check_for_red_dates == 1 :
        ax.annotate( str_curr, xy =( xpos_dates, -195 + count_add_pos ), size = 8, color ='red' )
    else :
        ax.annotate( str_curr, xy =( xpos_dates, -195 + count_add_pos ), size = 8 )
   
   

for d, e in zip( Crngtn_cycle_num, Crngtn_dates ) :
    if ( d != Crngtn_cycle_num[-2] ):
        ax.annotate( "cyc-"+str(d), xy =( e, 42 ), size = 8 )


HMFmt = mp.dates.DateFormatter('%b,%d')
ax.xaxis.set_major_formatter(HMFmt)
ax.xaxis.set_minor_locator(xtickMins_minor)


xticks(rotation=90,fontsize = 12 )
xlabel('Carrington cycle start dates ( 2012/2013 )', fontsize = 12)
ax.set_xticks( Crngtn_dates )
ylabel('Dst Index [nT]',fontsize = 12 )
xlim(date_st, date_end)
ylim(-200, 50)
title('24-hr Minimum Dst Vs Date / Carrington cycle', fontsize = 15 )



ax.grid( color = 'gray', linestyle = 'dashed' )
    
    
    
fig.savefig('/home/bharat/Desktop/DST-Crn2.pdf',orientation='portrait',format='pdf')
fig.savefig('/home/bharat/Desktop/DST-Crn2.jpeg',orientation='portrait',format='jpeg')
```
